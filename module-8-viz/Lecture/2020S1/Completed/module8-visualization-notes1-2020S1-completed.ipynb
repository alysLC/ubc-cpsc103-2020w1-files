## An Extra Note on Sorting

In the solution below, we take the information from the file with e-mail message counts in whatever order it comes. But, what if we wanted to sort it by year and month?

Python has a built-in function called `sorted` that is useful for this. Run `help(sorted)` in a Jupyter cell to see the help on this function, but the help is a bit hard to read! So, we'll demonstrate how it works here. The critical part is the `key` argument to `sorted` that lets us pass in a function (yes, pass in a function!) to indicate what to sort on.

To use it, first design a function that turns whatever you want to sort into a number (or string) that indicates its sort order. (So, if you just want to sort by one of the fields of a compound, you'd design a function that takes one of those compound values and returns just the field of interest.) Next, call sorted like this `sorted(LIST_HERE, key=FUNCTION_NAME_HERE)`.

Here's an example to sort a `List[EmailMonth]` by year/month:

```python
@typecheck
def get_time_value_for_month(em: EmailMonth) -> int:
    """
    returns a single number representing both the year and month from e-mail month
    by multiplying the year by 12 (converting it to months) and adding the month
    """
    #return 0  #stub
    #template from EmailMonth
    return em.year*12 + em.month

start_testing()
expect(get_time_value_for_month(EM1), 2004*12 + 8)
expect(get_time_value_for_month(EM2), 2018*12 + 2)
summary()

@typecheck
def sort_months_by_time(loem: List[EmailMonth]) -> List[EmailMonth]:
    """
    returns loem but ordered by increasing time (so by year and then if year ties by month)
    """
    #return []  #stub
    #return ...(loem)  #template (not using the List[EmailMonth] template!)
    return sorted(loem, key=get_time_value_for_month)

start_testing()

expect(sort_months_by_time(L3), [EM1, EM3, EM2])
expect(sort_months_by_time([EM2, EM3, EM1]), [EM1, EM3, EM2])
summary()
```

## A graphing problem with a bit of analysis

Design a function to graph the total number of emails in a year over time.

````python
@typecheck
def main_v2(filename: str) -> None:
    """
    Reads the file from given filename and plots a graph showing how the 
    number of emails sent per year has changed over time (returning None)
    """
    # Template from HtDAP, based on composition
    return plot_data_v2(read(filename), [2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019])

@typecheck
def plot_data_v2(loem: List[EmailMonth], years: List[int]) -> None:
    """
    Plots the number of emails sent per year in years (the parameter)
    """
    #return None  #stub
    # Template based on visualization
    
    y_values = get_email_counts_per_year(loem, years)
    
    # template from visualization
    # set the x-axis label, y-axis label, and plot title
    pyplot.xlabel('Year')
    pyplot.ylabel('Email Volume')
    pyplot.title('Number of Emails Sent Per Year')
    
    if loem != []:
        pyplot.axis([min(years) - 1, max(years) + 1, -100, max(y_values) + 100])
    
    # plot our data 
    line = pyplot.plot(years, y_values)

    # show the plot
    pyplot.show()
    
    return None


@typecheck
def get_email_counts_per_year(loem: List[EmailMonth], year_list: List[int]) -> List[int]:
    """
    Creates a list of ints that represent the number of emails sent each year in year_list
    """
    # return [] # List[int]
    
    # template from List[int]
    # acc will hold the list of emails for each year
    acc = [] # Type: List[int]
    
    for yr in year_list:
        acc.append(combine_counts_for_year(loem, yr))
    
    return acc

@typecheck
def combine_counts_for_year(loem: List[EmailMonth], year: int) -> int:
    """
    Finds the total number of emails sent for the given year.
    """
    
    # return 0 # Stub
    
    # template from List[EmailMonth]
    # total_count is the total number of emails seen so far from the EmailMonths in the year we are interested in
    total_count = 0 # Type: int
    
    for em in loem:
        # You might check this without a helper function, but it's a judgment call.. so we wrote one.
        if years_match(em, year):
            total_count = total_count + em.email_count
            
    return total_count

@typecheck
def years_match(em: EmailMonth, year: int) -> bool:
    """
    produce true if em is from the given year (and false otherwise)
    """
    #return False #stub
    
    # template based on EmailMonth
    return year == em.year
````
