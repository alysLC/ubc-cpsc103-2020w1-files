**_Note: Compared to the prep material, the table has one extra row, Sanders's Email has changed, and Warren's Choice has changed._**

Each row has:
+ a `Name` value, the respondent's name (a non-empty string, which may not contain an `@`)
+ an `Email` value, the respondent's e-mail address (a non-empty string, which must contain an `@`; **_if it begins with `@` like `@mochi`, then the actual e-mail address that should be stored as data is at `tweetbook.com` like `mochi@tweetbook.com`_**)
+ a timestamp (an integer greater than zero, where larger numbers mean later times)
+ a poll choice (which contains only letters, numbers, and spaces and must start with at least one letter) **_or `:NO VOTE:` to indicate that the person intentionally made no choice at all_**

**_Note: a voter may select multiple choices and so appear in multiple rows, unless they selected `:NO VOTE:`, in which case that must be their only row in the file._**

We are interested in the name, e-mail address, and choice. We represent them with these data definitions:

````python
Choice = Optional[str]      # NEW DATA DEFINITION
# interp. a poll choice containing only letters, numbers, and spaces and starting 
# with a letter, or None to indicate intentionally making no choice at all.
O0 = None
O1 = "unity"
O2 = "Est 1"

@typecheck    # template based on optional
def fn_for_choice(c: Choice) -> ...:
    if c is None:
        return ...
    else:
        return ...(c)

    
Vote = NamedTuple('Vote', [('name', str),
                           ('email', str),
                           ('choice', Choice)])  # CHANGED TYPE
# interp. a poll response, with the respondent's name (non-empty with no
# "@"), e-mail address (email, must have an "@" in it), and choice.
V1 = Vote("Joe Biden", "jb@gojoe.com", "Est 1")
V2 = Vote("Liz Warren", "ew@senate.gov", "unity")
V3 = Vote("T Gabbard", "tg@hawaii.us", None)  # not clearly necessary, but good to have an example with None

# template based on compound (3 fields) AND REFERENCE RULE # UPDATED!!
@typecheck
def fn_for_vote(v: Vote) -> ...:
    return ...(v.name,
               v.email,
               fn_for_choice(v.choice))   # UPDATED!!


# List[Vote]
# interp. a list of votes
LV1 = []
LV2 = [V1, V2]

# template based on arbitrary-sized and reference rule (Vote)
@typecheck
def fn_for_lov(lov: List[Vote]) -> ...:
    # description of accumulator
    acc = ...   # type: ...
    
    for v in lov:
        acc = ...(acc, fn_for_vote(v))
        
    return ...(acc)
````

````python
@typecheck
def read(filename: str) -> List[Vote]:
    """    
    reads information from the specified file and returns a list of the corresponding votes
    """
    # return []  #stub
    # Template from HtDAP

    # lov contains the result so far
    lov = [] # type: List[Vote]

    with open(filename) as csvfile:

        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            # you may not need to store all the rows, and you may need
            # to convert some of the strings to other types
            v = Vote(row[0], parse_email(row[1]), parse_choice(row[3]))
            lov.append(v)

    return lov

@typecheck
def parse_email(email: str) -> str:
    """
    Takes an email address that must have at least one letter and may be 
    a Tweetbook handle (i.e., starts with "@") and returns it as a proper
    e-mail address (unchanged unless it is a Tweetbook handle, in which case
    transformed from @HANDLE to HANDLE@tweetbook.com)
    """
    #return email  #stub
    #return ...(email)   #template
    
    # Since email must have at least one letter, we can safely access email[0].
    if email[0] == "@":
        return email[1:] + "@tweetbook.com"
    else:
        return email

@typecheck
def parse_choice(choice: str) -> Choice:
    """
    returns choice as a Choice, i.e., unchanged unless it is ":NO VOTE:", in
    which case None instead.
    
    (choice must be either a valid choice in a Piazza poll or ":NO VOTE:".)
    """
    #return None  #stub
    #return ...(choice)   #template
    
    if choice == ":NO VOTE:":
        return None
    else:
        return choice

start_testing()
expect(read('test_empty.csv'), [])

# It was handy to copy-paste the rows of the given test file down here
# to set up the test. We deleted them when we were done with them.
expect(read('test_given.csv'), [Vote('Joe Biden', 'jb@gojoe.com', 'Est 1'),
                                Vote('Bernard Sanders', 'bernie@tweetbook.com', 'health care'),
                                Vote('Joe Biden', 'jb@gojoe.com', 'unity'),
                                Vote('Liz Warren', 'ew@senate.gov', None)])
summary()

start_testing()
expect(parse_email("a@b"), "a@b")
expect(parse_email("pm@gov.ca"), "pm@gov.ca")
expect(parse_email("@"), "@tweetbook.com")
expect(parse_email("@boo"), "boo@tweetbook.com")
summary()

start_testing()
expect(parse_choice("Est 1"), "Est 1")
expect(parse_choice("anything"), "anything")
expect(parse_choice(":NO VOTE:"), None)
summary()
````
