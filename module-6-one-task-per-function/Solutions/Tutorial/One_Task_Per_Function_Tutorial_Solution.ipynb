## Problem 1

Design a function that takes a city name,  minimum grant amount, and a list of community grants and returns the community grants for the given city of at least the given minimum grant amount.

In particular, the **function MUST operate in two steps**. First, it will find the community grants for the given city, then it will find which of those community grants received at least the minimum grant amount.

Remember to follow the HtDF recipe and the helper rules.

**Thought question for pre-tuorial work:** Which helper rule applies because of the requirement to operate in two steps? *It's the composition rule because the work is divided into multiple, sequential, separate steps!*

```python
@typecheck
def filter_on_city_and_grant_amount(city: str, amt: int, locg: List[CommunityGrant]) -> List[CommunityGrant]:
    """
    return a list of grants from locg that are from the given city and of at least amt
    """
    #return [] #stub
    # template based on function composition
    
    # You aren't required to include a plan in comments, 
    # and your solution can be all in one line. But..
    # we like the way this looks and reads!
    
    
    # Plan:
    # 1) find the community grants in locg from city, call it city_grants
    # 2) find the community grants from city_grants with grant size at least amt, call it large_city_grants
    # 3) return large_city_grants
    
    city_grants = from_city_only(city, locg)
    large_city_grants = bigger_than_only(amt, city_grants)
    return large_city_grants
    
    # The alternate one-line version would be:
    #return bigger_than_only(amt, from_city_only(city, locg))
    
    # Can you see how that looks like the three lines above,
    # but with each variable name "substitude" for the expression
    # from which we got its value?

@typecheck
def bigger_than_only(amt: int, locg: List[CommunityGrant]) -> List[CommunityGrant]:
    """
    return a list of grants from locg that are of at least amt
    """
    # return [] #stub
    # template from List[CommunityGrant] with additional parameter
    
    # acc stores the result so far
    acc = [] # type: List[CommunityGrant]
    
    for cg in locg:
        if grant_bigger_than(amt, cg):
            acc.append(cg)
    
    return acc

@typecheck
def grant_bigger_than(i: int, cg: CommunityGrant) -> bool:
    """
    return True if cg has an amount greater than i and False otherwise
    """
    #return False #stub
    # template from CommunityGrant with additional parameter
    return cg.amt > i

@typecheck
def from_city_only(city: str, locg: List[CommunityGrant]) -> List[CommunityGrant]:
    """
    return a list of grants from locg that are from city
    """
    # return [] #stub
    # template from List[CommunityGrant] with additional parameter
    
    # acc stores the result so far
    acc = [] # type: List[CommunityGrant]
    
    for cg in locg:
        # It is perhaps a judgment call whether we need
        # a helper to check if cg is from city, but in
        # CPSC 103, we ask you to create this helper!
        if from_city(city, cg):
            acc.append(cg)
    
    return acc

@typecheck
def from_city(city: str, cg: CommunityGrant) -> bool:
    """
    return True if cg is from the given city, False otherwise
    """
    #return False #stub
    # template from CommunityGrant with additional parameter
    return cg.city == city

start_testing()

expect(filter_on_city_and_grant_amount("Burnaby", 20000, []), []) 
expect(filter_on_city_and_grant_amount("Burnaby", 20000, L1), [BARRACUDAS])
expect(filter_on_city_and_grant_amount("Burnaby", 2000, L1), [ARTS_COUNCIL, BARRACUDAS])
expect(filter_on_city_and_grant_amount("Burnaby", 15099, L1), [ARTS_COUNCIL, BARRACUDAS])
expect(filter_on_city_and_grant_amount("Richmond", 30000, L1), [YOUTH_MUSIC])
expect(filter_on_city_and_grant_amount("Richmond", 50000, L1), [])

expect(bigger_than_only(1, []), [])
expect(bigger_than_only(4999, L1), L1)
expect(bigger_than_only(30000, L1), [AGRIFAIR, AIRSHOW, BARRACUDAS, YOUTH_MUSIC])
expect(bigger_than_only(85000, L1), [])

expect(grant_bigger_than(3000, CHINESE_ORCHESTRA), True)
expect(grant_bigger_than(5000, CHINESE_ORCHESTRA), False)
expect(grant_bigger_than(10000, CHINESE_ORCHESTRA), False)
expect(grant_bigger_than(29499, BMX), True)
expect(grant_bigger_than(29500, BMX), False)
expect(grant_bigger_than(29501, BMX), False)

expect(from_city_only("Burnaby", []), []) 
expect(from_city_only("Burnaby", L1), [ARTS_COUNCIL, BARRACUDAS])
expect(from_city_only("Coquitlam", L1), [])
expect(from_city_only("Richmond", L1), [YOUTH_MUSIC, CHINESE_ORCHESTRA])

expect(from_city("Richmond", CHINESE_ORCHESTRA), True)
expect(from_city("Vancouver", CHINESE_ORCHESTRA), False)
expect(from_city("Abbotsford", BMX), True)
expect(from_city("North Vancouver", BMX), False)

summary()
```
